```python
# Cleaning coLumn genre_ id to genre_ names
# Removing [] from genre ids
tmdb_movies['genre _ids'] = tmdb_movies['genre_ids'].apply (lambda x: x.strip('[]').strip(""))
# Creating column genre List with the first value from genre ids
tmdb_movies['genre_list']= tmdb_movies['genre_ids'].apply (lambda x: x.split(',')[0])
# Creating a dictionary for genre names and replacing genre_ id with genre_ names,
# I googled info from tmab website source
# https://www.themoviedb.org/talk/5daf6eb@ae36680011d7e6ee
common_genrelist = {'action': '28','adventure': '12', 'animation': '16', 'comedy': '35', 'crime':'80',
'documentary': '99', 'drama': '18','family': '10751', 'fantasy': '14', 'history':'36','horror': '27', 'music': '10402',
'mystery': '9648',
'romance': '10749', 'scfiction': '878', 'tvmovie':
'10770',
'thriller': '53',
'war': '10752',
'western': '37'}
# Iterate over genre ids column to replace with genre name based on dictionary common genrelist
for k,v in common_genrelist.items() : tmdb_movies.loc[tmdb_movies['genre_list'].str.contains(v), 'genre_ list'] = k
tmdb_movies.head(10)

```

```python
def convert_dollars(dollar):
 return int(dollar.replace("$", "").replace(",", ""))
```

```python
tn_movie_budgets['production_budget'] = tn_movie_budgets['production_budget'].apply(convert_dollars)
```

```python
tn_movie_budgets['domestic_gross'] = tn_movie_budgets['domestic_gross'].apply(convert_dollars)
```

```python
tn_movie_budgets['worldwide_gross'] = tn_movie_budgets['worldwide_gross'].apply(convert_dollars)
```

```python
tn_movie_budgets.head()
```

```python
tn_movie_budgets['domestic_gross_mil']=tn_movie_budgets['domestic_gross']/1000000
tn_movie_budgets['worldwide_gross_mil']=tn_movie_budgets['worldwide_gross']/1000000
tn_movie_budgets['production_budget_mil']=tn_movie_budgets['production_budget']/1000000
```

```python
tn_movie_budgets['release_date'] = pd.to_datetime(tn_movie_budgets['release_date'])
tn_movie_budgets['day_of_week'] = tn_movie_budgets['release_date'].dt.day_name()
tn_movie_budgets['month'] = tn_movie_budgets['release_date'].dt.strftime('%B')
```

```python
fig, ax = plt.subplots(figsize = (12,8))
sns.lineplot(data=tn_movie_budgets, x="production_budget_mil", y="domestic_gross_mil")
plt.title('Budget versus Domestic Gross')
plt.xlabel('Production Budget in Millions')
plt.ylabel('Domestic Gross in Millions')
plt.show
```

```python
df_tmdb_and_tn = pd.concat([tn_movie_budgets, tmdb_movies], axis=1)

df_tmdb_and_tn
```

```python
df_tmdb_and_tn['day_of_week'] = df_tmdb_and_tn['day_of_week'].astype(str)
df_tmdb_and_tn['month'] = df_tmdb_and_tn['month'].astype(str)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_tmdb_and_tn['day_of_week'], df_tmdb_and_tn['domestic_gross_mil'], color='blue', alpha=0.5)
plt.title('Domestic Gross by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Domestic Gross')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)  # Add grid lines
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_tmdb_and_tn['month'], df_tmdb_and_tn['domestic_gross_mil'], color='orange', alpha=0.5)
plt.title('Domestic Gross by Month')
plt.xlabel('Month')
plt.ylabel('Domestic Gross')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)  # Add grid lines
plt.show()
```
